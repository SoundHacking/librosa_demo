```python
import librosa
audio, sample_rate = librosa.load("piano2.wav")
```

```python
print(sample_rate)
```

```python
import matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(audio)
ax.set(xlabel='time (samples)', ylabel='Sound',
       title='Audio')
fig.set_size_inches(12,3)
plt.show()
```

```python
tempo, beat_frames = librosa.beat.beat_track(y=audio, sr=sample_rate)
beat_times = librosa.frames_to_time(beat_frames, sr=sample_rate)
```

```python
y_harm, y_perc = librosa.effects.hpss(audio)
```

```python
y_harm, y_perc = librosa.effects.hpss(audio)
librosa.display.waveplot(y_harm, sr=sample_rate, alpha=0.25)
librosa.display.waveplot(y_perc, sr=sample_rate, color='r', alpha=0.5)
```

```python
import librosa.feature
mfcc = librosa.feature.mfcc(y=audio, sr=sample_rate)
mfcc_delta = librosa.feature.delta(mfcc)
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(mfcc_delta, ax=ax, x_axis='time')
ax.set(title=r'MFCC-$\Delta$')
ax.label_outer()
fig.colorbar(img, ax=[ax])
fig.set_size_inches(16,5)
```
